```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

#data file
covid_data_file = "Resources/200505COVID19MEXICO_RAW.csv"
# Read the data
covid_df = pd.read_csv(covid_data_file,encoding = "latin-1")

#to show all columns in head previews
pd.set_option('display.max_columns', None)

#function for converting codes to strings based on catalogs
def replaceCodes(catalogName,columnName):
    catalog_file = f"CATALOGOS/{catalogName}.csv"
    catalog_df = pd.read_csv(catalog_file,encoding = "latin-1")
    for index, row in catalog_df.iterrows():
        code = row['CLAVE']
        value = row['DESCRIPCION']
        covid_df[columnName] = covid_df[columnName].replace(code, value)
    print('processing',columnName)

#preview original data
covid_df.head()
```

```python
# Bernardo Jardon
columns_raw = [("ORIGEN","ORIGEN"),("SECTOR","SECTOR"),("ENTIDADES","ENTIDAD_UM"),("SEXO","SEXO"),("ENTIDADES","ENTIDAD_NAC"),("ENTIDADES","ENTIDAD_RES"),("MUNICIPIOS","MUNICIPIO_RES")]
for catcol in columns_raw:
    replaceCodes(catcol[0],catcol[1])
#check to see if replace was successfull
covid_df.head()
```

```python
# Odín Efrén Del Toro Reyna
columns_raw = [("TIPO_PACIENTE","TIPO_PACIENTE"),("SI_NO","INTUBADO"),("SI_NO","NEUMONIA"),("NACIONALIDAD","NACIONALIDAD"),("SI_NO","EMBARAZO"),("SI_NO","HABLA_LENGUA_INDIG"),("SI_NO","DIABETES")]
for catcol in columns_raw:
    replaceCodes(catcol[0],catcol[1])
#check to see if replace was successfull
covid_df.head()
```

```python
# Miguel Angel Bonachea Soto
columns_raw = [("SI_NO","EPOC"),("SI_NO","ASMA"),("SI_NO","INMUSUPR"),("SI_NO","HIPERTENSION"),("SI_NO","OTRA_COM"),("SI_NO","CARDIOVASCULAR"),("SI_NO","OBESIDAD"),("SI_NO","RENAL_CRONICA")]
for catcol in columns_raw:
    replaceCodes(catcol[0],catcol[1])
#check to see if replace was successfull 
covid_df.head()
```

```python
# Pablo Arroyo
# TABAQUISMO
# OTRO_CASO
# RESULTADO
# MIGRANTE
# PAIS_NACIONALIDAD
# PAIS_ORIGEN
# UCI
columns_raw = [("SI_NO","TABAQUISMO"),("SI_NO","OTRO_CASO"),("RESULTADO","RESULTADO"),("SI_NO","MIGRANTE"),("NACIONALIDAD","PAIS_NACIONALIDAD"),("SI_NO","UCI")]
for catcol in columns_raw:
    replaceCodes(catcol[0],catcol[1])

covid_df['PAIS_ORIGEN'] = covid_df['PAIS_ORIGEN'].replace('99', 'Mexico')
#check to see if replace was successfull 
covid_df.head()
```

```python
# export clean data to .csv
covid_df.to_csv('COVID19MX-CLEAN.csv', index=False)
```
